```python
import pandas as pd

s3_path = "s3://sales-project-dataanalysis/superstore.csv"

df = pd.read_csv(s3_path, storage_options={"anon": False})

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df = df.where(pd.notnull(df), None)
```

```python
df.dropna(inplace=True)
df
```

```python
# Columns to clean
str_cols = ['Product.Name','Customer.Name','Sub.Category','Market','Market2']

# Strip spaces and truncate to 500 characters
for col in str_cols:
    df[col] = df[col].astype(str).str.strip().str.slice(0, 500)
df
```

```python
df['Order.Date'] = pd.to_datetime(df['Order.Date'], errors='coerce')
df['Ship.Date'] = pd.to_datetime(df['Ship.Date'], errors='coerce')
```

```python
# Remove exact duplicate rows
df = df.drop_duplicates()

# Optionally, remove duplicate Row.IDs keeping the first occurrence
df = df.drop_duplicates(subset=['Row.ID'], keep='first')
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS sales (
    `Row.ID` INT PRIMARY KEY,
    `Order.ID` VARCHAR(50),
    `Order.Date` DATE,
    `Ship.Date` DATE,
    `Customer.ID` VARCHAR(50),
    `Customer.Name` VARCHAR(100),
    Segment VARCHAR(50),
    Country VARCHAR(50),
    City VARCHAR(50),
    State VARCHAR(50),
    Region VARCHAR(50),
    Category VARCHAR(50),
    `Sub.Category` VARCHAR(50),
    `Product.ID` VARCHAR(50),
    `Product.Name` VARCHAR(100),
    Sales FLOAT,
    Quantity INT,
    Discount FLOAT,
    Profit FLOAT,
    `Order.Priority` VARCHAR(20),
    `Ship.Mode` VARCHAR(50),
    `Shipping.Cost` FLOAT,
    `记录数` INT,
    Year INT,
    Market VARCHAR(50),
    Market2 VARCHAR(50),
    weeknum INT
);
"""

cursor.execute(create_table_query)
connection.commit()
print("Table created successfully!")
```

```python
# --- Step 2: Row-by-row insertion using iterrows ---
for _, row in df.iterrows():
    sql = """
    INSERT INTO sales 
    (`Row.ID`, `Order.ID`, `Order.Date`, `Ship.Date`, `Customer.ID`, `Customer.Name`,
     Segment, Country, City, State, Region, Category, `Sub.Category`, `Product.ID`,
     `Product.Name`, Sales, Quantity, Discount, Profit, `Order.Priority`, `Ship.Mode`,
     `Shipping.Cost`, `记录数`, Year, Market, Market2, weeknum)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    """
    # Added one %s above (total 27)
    
    values = (
        row['Row.ID'], row['Order.ID'], row['Order.Date'], row['Ship.Date'], row['Customer.ID'], row['Customer.Name'],
        row['Segment'], row['Country'], row['City'], row['State'], row['Region'], row['Category'], row['Sub.Category'],
        row['Product.ID'], row['Product.Name'], row['Sales'], row['Quantity'], row['Discount'], row['Profit'],
        row['Order.Priority'], row['Ship.Mode'], row['Shipping.Cost'], row['记录数'], row['Year'], row['Market'],
        row['Market2'], row['weeknum']
    )
    
    cursor.execute(sql, values)
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby('Category')['Sales'].sum()
category_sales.plot.pie(autopct='%1.1f%%', figsize=(6,6), startangle=90)
plt.title("Sales Distribution by Category")
plt.ylabel("")
plt.show()
```

```python
top_products = df.groupby('Product.Name')['Profit'].sum().sort_values(ascending=False).head(10)
top_products.plot.barh(color='skyblue', figsize=(8,6))
plt.xlabel("Profit")
plt.ylabel("Product")
plt.title("Top 10 Products by Profit")
plt.gca().invert_yaxis()
plt.show()
```

```python
df['YearMonth'] = df['Order.Date'].dt.to_period('M')
monthly_sales = df.groupby('YearMonth')['Sales'].sum()
monthly_sales.plot(linewidth=2, color='green', marker='o', figsize=(10,5))
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()
region_sales.plot.bar(color='orange', figsize=(8,6))
plt.ylabel("Sales")
plt.title("Total Sales by Region")
plt.show()
```

```python
shipmode_sales = df.groupby('Ship.Mode')['Sales'].sum()
shipmode_sales.plot.pie(autopct='%1.1f%%', figsize=(6,6), startangle=140)
plt.title("Sales Distribution by Ship Mode")
plt.ylabel("")
plt.show()
```
